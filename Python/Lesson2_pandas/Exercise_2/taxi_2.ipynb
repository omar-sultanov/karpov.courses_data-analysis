### Следующая задача: выяснить, из какого района было совершено наибольшее количество поездок за весь период. Для начала посчитайте общее количество поездок (pickups) без группировки.

```python
taxi.pickups.sum()
```

```python
taxi.groupby('borough').pickups.sum()
```

### Сохраните название района с наименьшим числом поездок в переменную min_pickups, применив подходящий метод.

```python
min_pickups = taxi.groupby('borough').pickups.sum().idxmin()
```

```python
min_pickups
```

### Продолжим изучение данных и посмотрим на число поездок в выходные дни. Сгруппируйте данные по двум признакам: району города и является ли день выходным (колонки borough и hday). Сравните среднее число поездок, и выберите районы, из которых по праздникам в среднем поступает больше заказов, чем в обычные дни. 

```python
taxi.groupby(['borough','hday']).pickups.mean()
```

### Для каждого района посчитайте число поездок по месяцам. Отсортируйте полученные значения по убыванию и сохраните результирующий датафрейм в pickups_by_mon_bor. Обратите внимание, что итоговый датасет должен состоять из 3-х колонок: pickup_month, borough, pickups.

```python
pickups_by_mon_bor = taxi.groupby(['pickup_month','borough'], as_index=False).aggregate({'pickups':'sum'}).sort_values('pickups', ascending=False)
```

```python
pickups_by_mon_bor
```

### Напишите функцию temp_to_celcius, которая получает на вход колонку с температурой в °F и возвращает значения, переведённые в градусы Цельсия.

```python
def temp_to_celcius(temp):
    return ((temp - 32)*5.0)/9.0
```

```python
taxi['temp'][:5]
```

```python
taxi['temp_C'] = temp_to_celcius(taxi['temp'])
```

```python
df = df.rename(columns={'Номер':'number', 'Дата создания':'date_of_created', 'Дата оплаты':'date_of_payment', 'Title':'title', 'Статус':'status',
       'Заработано':'money', 'Город':'city', 'Платежная система':'payment_system'})
```

```python
df.head()
```

```python
all_money= df.money.sum()
```

```python
all_money
```

### Группировка, агрегация, сортировка данных

```python
money_by_city= df\
.groupby(['title','city'], as_index=False)\
.agg({'money':'sum'})\
.sort_values('money', ascending=False)
```

```python
money_by_city
```

```python
money_by_city.to_csv('money_by_city.csv', index=False, encoding='windows-1251')
```

```python
money_by_status= df\
.query("status =='Завершен'",  inplace=False)\
.groupby(['title'], as_index=False)\
.agg({'money':'sum', 'number':'count'})\
.sort_values('money', ascending=False)\
.rename(columns={'number':'success_order'})
```

```python
money_by_status
```

### Автоматизация скучной работы

```python
current_date = datetime.datetime.today().strftime('%d-%m-%Y')
```

```python
report_file_name = 'report_{}.csv'.format(current_date)
```

```python
if int(money_by_status.money.sum()) == int(all_money):
    print('OK')
    money_by_status.to_csv(report_file_name, index=False)
else:
    print('ERROR')
```

```python
import pandas as pd
import numpy as np
import sqlite3
```

```python
con = sqlite3.connect('db')
df = pd.read_csv('../german_credit_augmented.csv')
```

```python
df.to_sql('german_credit', con, index=False, if_exists='replace')
```

```python
sql = '''select * from german_credit as t'''
```

```python
def select(sql):
    return pd.read_sql(sql, con)
```

```python
select(sql)
```

```python
sql = '''select t.sex, t.housing, t.purpose,
count(t.purpose) as cnt,
avg(t.credit_amount) as credit_amount_mean
 from german_credit as t
 group by t.sex, t.housing, t.purpose
 '''
```

```python
select(sql)
```

```python
df.groupby(['sex','housing', 'purpose'], as_index=False).agg({'credit_amount':'mean'})
```

```python
sql = '''select count(t.checking_account), count(*) from german_credit as t'''
```

```python
select(sql)
```

```python
sql = '''select t.checking_account,
count(*) as cnt,
avg(t.credit_amount) as credit_amount_mean
 from german_credit as t
 group by t.checking_account
 '''
```

```python
select(sql)
```

```python
df.groupby('checking_account', dropna=False)['credit_amount'].count()
```

```python
sql = '''select t.checking_account,  
sum(case when t.checking_account is null then 1 else 0 end) as is_null,
count(case when t.checking_account is null then 1 else null end) as is_null2
 from german_credit as t'''
```

```python
select(sql)
```

```python
# null and NaN не расчитает функция count()
```

```python
sql = '''select t.checking_account,  
coalesce(t.checking_account, 'null')
 from german_credit as t'''
```

```python
select(sql)
```

```python
sql = '''select
        case 
        when t.credit_amount < 1000 then '1.<1000' 
        when t.credit_amount < 2000 then '2.<2000'
        when t.credit_amount < 3000 then '3.<2000'
        when t.credit_amount >= 3000 then '2.>= 3000'
        else  t.credit_amount end as credit_amount_bin,
        count(*) as cnt 
        from german_credit as t
        group by credit_amount_bin'''
```

```python
select(sql)
```
